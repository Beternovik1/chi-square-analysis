```python
import pandas as pd
import numpy as np
import sys
sys.path.append("../")
from src.stats_utils import analyze_goodness_of_fit

# Cargar las frecuencias de la simulación
df = pd.read_csv("../data/5_bondad_ajuste_confusion_teorica.csv")

# Extraer vectores de la tabla proporcionada
categorias = df['category'].values
observados = df['observed'].values
esperados = df['expected_count'].values

# Ejecutar la prueba de Bondad de Ajuste
chi2, p_val = analyze_goodness_of_fit(observados, esperados)

# Display analítico
comparacion = pd.DataFrame({
    'Observados (O)': observados,
    'Esperados (E)': esperados,
    'Proporción Teórica': df['expected_prop'].values
}, index=categorias)

print("=== MATRIZ DE COMPARACIÓN DE ERRORES ===")
print(comparacion)

print("\n=== MÉTRICAS ESTADÍSTICAS ===")
print(f"Estadístico Chi-cuadrada: {chi2:.4f}")
print(f"Valor p: {p_val:.4e}")

# Evaluación de la Hipótesis
alpha = 0.05
if p_val < alpha:
    print(f"\n[CONCLUSIÓN] Se rechaza H0 (p < {alpha}). El modelo NO funciona según lo simulado.")
else:
    print(f"\n[CONCLUSIÓN] Se retiene H0 (p >= {alpha}). Las distribuciones de error se ajustan al modelo teórico.")
```
