```python
import pandas as pd
import sys
sys.path.append("../")
from src.stats_utils import analyze_independence

# Load the dataset
df = pd.read_csv("../data/4_homogeneidad_arquitectura_por_empresa.csv")

# Execute the independence/homogeneity test
col_company = 'company'
col_architecture = 'architecture'

observed, expected, chi2, p_val, cramer_v = analyze_independence(df, col_company, col_architecture)

# Display results analytically
print("=== MATRIZ DE FRECUENCIAS OBSERVADAS ===")
print(observed)
print("\n=== MATRIZ DE FRECUENCIAS ESPERADAS ===")
print(expected)
print("\n=== MÉTRICAS ESTADÍSTICAS ===")
print(f"Estadístico Chi-cuadrada: {chi2:.4f}")
print(f"Valor p: {p_val:.4e}")
print(f"V de Cramer (Tamaño del efecto): {cramer_v:.4f}")

# Hypothesis Evaluation
alpha = 0.05
if p_val < alpha:
    print(f"\n[CONCLUSIÓN] Se rechaza H0 (p < {alpha}). Las distribuciones de preferencia NO son homogéneas entre empresas.")
else:
    print(f"\n[CONCLUSIÓN] Se retiene H0 (p >= {alpha}). Las distribuciones de preferencia son HOMOGÉNEAS entre empresas.")
```
