```python
import pandas as pd
import sys
sys.path.append("../")
from src.stats_utils import analyze_independence

df = pd.read_csv("../data/3_independencia_falla_vs_sensor.csv")


# Execute the independence test
col_sensor = 'sensor_type'
col_failure = 'failure_type'

observed, expected, chi2, p_val, cramer_v = analyze_independence(df, col_sensor, col_failure)

# Display results analytically
print("=== MATRIZ DE FRECUENCIAS OBSERVADAS ===")
print(observed)
print("\n=== MATRIZ DE FRECUENCIAS ESPERADAS ===")
print(expected)
print("\n=== MÉTRICAS ESTADÍSTICAS ===")
print(f"Estadístico Chi-cuadrada: {chi2:.4f}")
print(f"Valor p: {p_val:.4e}")
print(f"V de Cramer (Tamaño del efecto): {cramer_v:.4f}")

# Hypothesis Evaluation
alpha = 0.05
if p_val < alpha:
    print(f"\n[CONCLUSIÓN] Se rechaza H0 (p < {alpha}). El tipo de falla DEPENDE del tipo de sensor.")
else:
    print(f"\n[CONCLUSIÓN] Se retiene H0 (p >= {alpha}). El tipo de falla es INDEPENDIENTE del tipo de sensor.")
```
