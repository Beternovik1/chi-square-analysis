### Exercise 2: Algorithmic Bias Detection in Recommendation Systems

This analysis evaluates if a user's acceptance of a recommendation is independent of their age group, and measures the effect size of that relationship.

#### 1. Chi-Square Test of Independence

We test the null hypothesis ($H_0$) that `age_group` and `accepted` are strictly independent.
$$\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$
Where:
* $O_{ij}$ = Observed frequency in the contingency table
* $E_{ij}$ = Expected frequency assuming true independence

#### 2. Cramer's V (Effect Size)


[Image of Cramer's V formula]

Because the Chi-Square statistic scales with sample size, we calculate Cramer's V to determine the actual strength of the association on a normalized scale from 0 to 1.
$$V = \sqrt{\frac{\chi^2}{n \cdot \min(r-1, c-1)}}$$
Where:
* $n$ = Total number of observations
* $r, c$ = Number of rows and columns in the contingency table
